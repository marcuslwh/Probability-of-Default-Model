### General Preprocessing

```python
loan_data['emp_length'].unique()
```

```python
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('\+ years', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('< 1 year', '0')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' years', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' year', '')
loan_data['emp_length_int'].unique()
```

```python
loan_data['emp_length_int'].head(100)
```

```python
type(loan_data['emp_length_int'][0])
```

```python
loan_data['emp_length_int'] = pd.to_numeric(loan_data['emp_length_int'], downcast='integer')
```

```python
type(loan_data['emp_length_int'][0])
```

```python
loan_data['earliest_cr_line'].head(100)
```

```python
type(loan_data['earliest_cr_line'][0])
```

```python
loan_data['earliest_cr_line_date'] = pd.to_datetime(loan_data['earliest_cr_line'], format='%b-%Y')
loan_data['earliest_cr_line_date'].head(100)
```

```python
type(loan_data['earliest_cr_line_date'][0])
```

```python
loan_data['mths_since_earliest_cr_line'] = round(pd.to_numeric((pd.to_datetime('2021-10-01') - loan_data['earliest_cr_line_date']) / np.timedelta64(1, 'M')))
loan_data['mths_since_earliest_cr_line'].describe()
```

```python
loan_data['term'].head(100)
```

```python
loan_data['term'].describe()
```

```python
loan_data['term_int'] = pd.to_numeric(loan_data['term'].str.replace(' months', ''))
loan_data['term_int'].head(100)
```

```python
type(loan_data['term_int'][0])
```

```python
loan_data['issue_d'].head(50)
```

```python
loan_data['issue_d'].unique()
```

```python
loan_data['issue_d_date'] = pd.to_datetime(loan_data['issue_d'], format='%b-%Y')
loan_data['mths_since_issue_d'] = round(pd.to_numeric((pd.to_datetime('2021-10-01') - loan_data['issue_d_date']) / np.timedelta64(1, 'M')))
loan_data['mths_since_issue_d'].describe()
```

```python
loan_data_dummies = pd.get_dummies(loan_data[['grade', 'sub_grade', 'home_ownership', 'verification_status', 'loan_status', 'purpose', 'addr_state', 'initial_list_status']], prefix_sep=':')
```

```python
loan_data = pd.concat([loan_data, loan_data_dummies], axis=1)
loan_data.columns.values
```
